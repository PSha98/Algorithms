### Singular Value Decomposition Algorithm
Kernel-SageMath

Theorem- Let A be a real mxn matrix. Then $\exists$ orthogonal matrices U and V such that 

A = U$\sum V^T$

where U is m x n orthogonal matrix and V is n x n orthogonal matrix and $\sum$ is mxn diagonal matrix whose diagonal entries are non-negative and are arranged in a non-increasing order. The number of non-zero entries in $\sum$ is the rank of A 

Let $U = [u_1,u_2.....u_m], V = [v_1,v_2.......v_n]$ and 

\begin{equation}
\Sigma = \begin{bmatrix} 
    \sigma_{1} & 0 & \dots & 0 & 0 \dots & 0 \\
     0 & \sigma_2 & \dots & 0 & 0 \dots & 0 \\
    \vdots & \vdots & \ddots & 0 & 0 \ddots & \vdots \\
    0 & 0 & \dots & \sigma_r & 0 \dots & 0 \\
\hline
0 & 0 & \dots & 0 & 0 \dots & 0\\
 \vdots & \vdots & \ddots & \vdots & \vdots \ddots & \vdots \\
0 & 0& \dots & 0 & \dots & 0
\end{bmatrix}
\end{equation}

1. We have $\sigma_1 \geq\sigma_2\geq\sigma_3........\geq\sigma_r$ are singular values of A. Here r is the rank of A.

2. $A^TA$ and $AA^T$ both are symmetric and hence diagonalizable

3. $x^TA^TAx$ = $\langle A^TAx,x\rangle$ = $\langle Ax,Ax \rangle$ = $\Vert Ax \Vert^2\geq0$ that is $A^TA$ is symmetric and non-negative definite. Similarly $A^TA$ is also symmetric and non-negative definite.

4. If A = U$\sum$$V^T$, then 

$A^TA$ = $(U \sum V^T)^T (U \sum V^T)$ = $V(\sum^T\sum)V^T$

that is 

$V^T(A^TA)V = \sum^T\sum = diag(\sigma_1^2,......\sigma_r^2,0,.....0)_mxn$

Thus columns of V are eigenvectors of $A^TA$ with eigenvalues $\sigma_1^2.....\sigma_r^2,0....,0$
$AA^T$ = ($U\sum V^T)(U\sum V^T)^T$ = $U(\sum\sum^T)U^T$
That is 

$U^T(AA^T)U = \sum^T\sum  = diag(\sigma_1^2....,\sigma_r^2,0,...0)_nxn $ 
The columns of U are eigenvectors of $AA^T$

```python
A = matrix([[4,-2],[-2,3]])
U,S,VT = A.change_ring(RDF).SVD()
S
```

```python
var('t')
cir = vector([cos(t),sin(t)])
c1 = parametric_plot(cir,(t,0,2*pi),figsize=4)
c1
```

```python
e1 = vector([1,0])
e2 = vector([0,1])
svd1=parametric_plot(VT*cir,(t,0,2*pi),figsize=4)
svd1 = svd1+plot(VT*e1,color='red')+plot(VT*e2,color='green')
svd1
```

```python
svd2=parametric_plot(S*VT*cir,(t,0,2*pi),figsize=4)
svd2 = svd2+plot(S*VT*e1,color='red')+plot(S*VT*e2,color='green')
svd2
```

```python
svd3 = parametric_plot(U*S*VT*cir,(t,0,2*pi),figsize=4)
svd3=svd3+plot(U*S*VT*e1,color='red')+plot(U*S*VT*e2,color='green')
svd3
```
